```python
# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
# create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
# for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
# should include the current day.
data = {
    "A": [3,8,6,2,9],
    "B": [5,2,9,3,1],
    "C": [1,7,4,5,2]
}
```

```python
import pandas as pd

def calculate_moving_average(df):
    """
    Adds a new column 'MovingAverage' to the DataFrame that contains
    the 7-day moving average of the 'Sales' column.
    """
    # Ensure the DataFrame is sorted by 'Date' (important for time-series calculations)
    df = df.sort_values('Date')

    # Calculate the 7-day moving average and add it as a new column
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

# Example DataFrame
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sales': [200, 220, 210, 240, 250, 260, 270, 280, 300, 310]
}
df = pd.DataFrame(data)

# Call the function
df = calculate_moving_average(df)

# Display the DataFrame
print(df)
```

```python
Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
```

```python
import pandas as pd

def add_weekday_column(df):
    """
    Adds a new column 'Weekday' to the DataFrame that contains the weekday name
    corresponding to each date in the 'Date' column.
    """
    # Ensure that the 'Date' column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Create the 'Weekday' column by extracting the weekday name
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df

# Example DataFrame with a 'Date' column
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
}
df = pd.DataFrame(data)

# Call the function to add the 'Weekday' column
df = add_weekday_column(df)

# Display the updated DataFrame
print(df)
```

```python
# Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
# function to select all rows where the date is between '2023-01-01' and '2023-01-31'.
data = {
    'Date': ['2023-01-01', '2023-01-15', '2023-02-01', '2023-01-25', '2023-02-10'],
}
df = pd.DataFrame(data)
start_date = '2023-01-01'
end_date = '2023-01-31'

# Filter the data frame using boolean indexing
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
filtered_df
```
