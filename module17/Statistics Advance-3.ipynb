```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard
# deviation.
def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    from math import sqrt
    standard_error = sample_std_dev / sqrt(sample_size)
    return sample_mean, standard_error
estimate_population_mean(5,10,36)
```

```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.
# Ho="Reject the null hypothesis: There is a significant difference between the two population means."
#  H1="Fail to reject the null hypothesis: No significant difference between the two population means."
from scipy.stats import ttest_ind
def hypothesis_test_two_means(sample1, sample2):
    t_stat, p_value = ttest_ind(sample1, sample2)
    return t_stat, p_value
sample1 = [5.1, 4.9, 5.0, 4.8, 5.2]
sample2 = [5.5, 5.6, 5.7, 5.4, 5.8]    
hypothesis_test_two_means(sample1, sample2)
# let alpha =0.05
# p value is less than alpha then reject the null hypothesis 
```

```python
# Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
# parameter set to 10.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df=10)

plt.plot(x, y, label="t-distribution (df=10)")
plt.title("Student's t-distribution")
plt.xlabel("t")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=True)
    return t_stat, p_value
```
