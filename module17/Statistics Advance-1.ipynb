```python
# Q3. Write a Python function to calculate the probability density function of a normal distribution with
# given mean and standard deviation at a given point.    
import math
import numpy as np
import matplotlib.pyplot as plt
def normal_pdf(x, mean, std_dev):
    return (1 / (math.sqrt(2 * np.pi) * std_dev)) * np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))

# Example Usage
x= np.linspace(-5,5,100)

mean = 0
std_dev = 1
n=normal_pdf(x, mean, std_dev)
plt.plot(n)
```

```python
# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
# and plot a histogram of the results using matplotlib.
import numpy as np
import matplotlib.pyplot as plt

# Generate random sample
n, p = 10, 0.4
sample = np.random.binomial(n, p, 1000)

# Plot histogram
plt.hist(sample, bins=10, edgecolor='k', alpha=0.7)
plt.title("Binomial Distribution Histogram")
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
# with given mean at a given point.
import numpy as np
import matplotlib.pyplot as plt
import math
def poisson_cdf(mean, x):
    cdf = 0
    cdf_values = []
    for k in range(0, x + 1):
        cdf += (math.exp(-mean) * (mean ** k)) / math.factorial(k)
        cdf_values.append(cdf)
    return cdf_values

def plot_poisson_cdf(mean, max_x):
    x_values = list(range(max_x + 1))
    cdf_values = poisson_cdf(mean, max_x)
    
    plt.plot(x_values, cdf_values, marker='o', color='b', label=f"Poisson CDF (λ={mean})")
    plt.show()

plot_poisson_cdf(mean=5, max_x=20)
```
